##First Let's understand the basic Catalog + Volume Feature of Databricks (We are learning how to build & use Databricks DATALAKE)
#####Filesystem Hierarchy of Volume in Databricks (DBFS)?
Catalog -> /OurWorkspace/catalog/schema(database)/volume/folder/data files

#####Tables Hierarchy of Databricks?

Catalog -> /OurWorkspace/catalog/schema(database)/tables/data(dbfs filesystem/some other filesystems)


```python
%sql

create catalog if not exists catalog_usecase_partice;
create schema if not exists catalog_usecase_partice.usecase_partice_schema;
create volume if not exists catalog_usecase_partice.usecase_partice_schema.usercase_volmue;
use catalog_usecase_partice.usecase_partice_schema;
```

```python
df1=spark.read.csv("dbfs:/Volumes/catalog_usecase_partice/usecase_partice_schema/usercase_volmue/custs_header",header=True)
df1.show(10)
write_df1=df1.write.csv(path="/Volumes/catalog_usecase_partice/usecase_partice_schema/traget_file",sep='|',header=True,mode="overwrite")


```

```python
df2=spark.read.csv("dbfs:/Volumes/catalog_usecase_partice/usecase_partice_schema/traget_file",header=True)
df2.show(10)
write_df1=df1.write.csv(path="/Volumes/catalog_usecase_partice/usecase_partice_schema/usercase_volmue/Rewrite",sep=',',header=True,mode="overwrite")
```
